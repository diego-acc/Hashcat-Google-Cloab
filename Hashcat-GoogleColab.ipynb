# Práctica



## 1.   Straight Mode



```python
#Archivo raw-md5.hashes3
!hashcat -a 0 -m 0 Hashes/raw-md5.hashes3.txt Dicc/500_passwords.txt -o Results/pass_straight_3.txt --potfile-disable
!echo \ && cat Results/pass_straight_3.txt
```

```python
!hashcat -m 0 -a 0 -j '$!' Hashes/raw-md5.hashes3.txt Dicc/500_passwords.txt -o Results/pass_straight2_3.txt --potfile-disable
!echo \ && cat Results/pass_straight2_3.txt
```

```python
#Archivo raw-md5.hashes10
!hashcat -a 0 -m 0 Hashes/raw-md5.hashes10.txt Dicc/rockyou.txt -o Results/pass_straight_10.txt --potfile-disable
```

```python
!cat Results/pass_straight_10.txt
```


## 2.   Combinator Mode

```python
#Archivo raw-md5.hashes3
!hashcat -a 1 -m 0 Hashes/raw-md5.hashes3.txt Dicc/500_passwords.txt Dicc/passwords.txt -o Results/pass_comb_3.txt --potfile-disable
```

```python
!cat Results/pass_comb_3.txt
```

```python
!hashcat -a 1 -m 0 -j '$-' Hashes/raw-md5.hashes3.txt Dicc/500_passwords.txt Dicc/passwords.txt -o Results/pass_comb2_3.txt --potfile-disable
!echo \ && cat Results/pass_comb2_3.txt
```

```python
#Archivo raw-md5.hashes10
!hashcat -a 1 -m 0 Hashes/raw-md5.hashes10.txt Dicc/rockyou.txt Dicc/passwords.txt -o Results/pass_comb_10.txt --potfile-disable
```


## 3.   Hybrid Wordlist + Mask

```python
#Archivo raw-md5.hashes10
!hashcat -a 6 -m 0 Hashes/raw-md5.hashes3.txt Dicc/500_passwords.txt 20?d?d -o Results/pass_word_mask_3.txt --potfile-disable
!echo \ && cat Results/pass_word_mask_3.txt
```

```python
#Archivo raw-md5.hashes10
!hashcat -a 6 -m 0 Hashes/raw-md5.hashes10.txt Dicc/rockyou.txt 20?d?d -o Results/pass_word_mask_10.txt --potfile-disable
```


## 4.  Hybrid Mask + Wordlist

```python
#Archivo raw-md5.hashes10
!hashcat -a 7 -m 0 Hashes/raw-md5.hashes3.txt ?l?l?l?l Dicc/500_passwords.txt -o Results/pass_mask_word_3.txt --potfile-disable
!echo \ && cat Results/pass_mask_word_3.txt
```

```python
#Archivo raw-md5.hashes10
!hashcat -a 7 -m 0 Hashes/raw-md5.hashes10.txt ?l?l?l?l Dicc/passwords.txt -o Results/pass_mask_word_10.txt --potfile-disable
```


## 5.  Brute-force

```python
#Archivo raw-md5.hashes10
!hashcat -a 3 -m 0 Hashes/raw-md5.hashes3.txt ?l?u?d?s -o Results/pass_bruteforce_3.txt --potfile-disable
!echo \ && cat Results/pass_bruteforce_3.txt
```

```python
#Archivo raw-md5.hashes10
!hashcat -a 3 -m 0 Hashes/raw-md5.hashes10.txt ?l?u?d?s -o Results/pass_bruteforce_10.txt --potfile-disable
!echo \ && cat Results/pass_bruteforce_10.txt
```
